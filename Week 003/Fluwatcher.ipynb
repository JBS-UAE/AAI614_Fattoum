```python
#Import needed libraries
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd

def get_fluwatch_table(url, table_id):
    #Open the url
    html = urlopen(url)
    bs = BeautifulSoup(html, 'html.parser')
    
    #Locate table by ID
    table = bs.find("table", {"id": table_id})

    #Extract header
    headers = [header.text.strip() for header in table.find("thead").find("tr").find_all("th")]

    #Extract rows
    rows = []
    for row in table.find("tbody").find_all("tr"):
        cells = [cell.text.strip() for cell in row.find_all("td")]
        rows.append(cells)

    #Create dataframe
    df = pd.DataFrame(rows, columns=headers)
    return df

#URL and table ID for FluWatch
url = "https://www.canada.ca/en/public-health/services/surveillance/respiratory-virus-detections-canada/2018-2019/respiratory-virus-detections-isolations-week-01-ending-january-5-2019.html"
tid = "t2" #Table ID

#Scrape
data = get_fluwatch_table(url, tid)

#Display the dataframe
print(flu_data)

# Save dataframe to a CSV file
data.to_csv("Table02.csv", index=False)
```
